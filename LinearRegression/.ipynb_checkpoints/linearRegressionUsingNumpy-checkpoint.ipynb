```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, '.')
plt.xlabel('x1')
plt.ylabel('y')
plt.axis([0, 2, 0, 15])
plt.title('Randomly generated linear dataset')
```

```python
X_b = np.c_[np.ones((100, 1)), X] 
# np.c_ convert arrays in one on the second, third etc axis

theta_best0 = np.linalg.inv(X_b.T.dot(X_b))
# np.linealg.inv compute the (multiplicative) inverse of a matrix.
# dot product of two arrays. Specifically
theta_best1 = theta_best0.dot(X_b.T)

theta_best2 = theta_best1.dot(y)

theta_best = theta_best2
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, ".")
plt.axis([0, 2, 0, 15])
plt.show()
```
